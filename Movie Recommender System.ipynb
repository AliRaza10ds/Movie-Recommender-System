```python
data_movies=pd.read_csv('tmdb_5000_movies.csv')
data_movies.head()
```

```python
final_data=data_movies.merge(data,on='title')
```

## Dropping the non-useful columns
- There is no need of budget column in recommendation
- No use of Movie's Homepage
- Also dropping original language and original title column
- Making content based recommendation system so no need of popularity and production companies and production_countries
- No need of revenue and release_date
- and other non-useful columns 

```python
final_data=final_data[['movie_id','title','overview','genres','keywords','cast','crew']]
final_data.head()
```

```python
final_data.isnull().sum()
```

```python
## overview column has 3 null values so removing null values 
final_data.dropna(inplace=True)
```

```python
final_data.duplicated().sum()
```

## Data Preprocessing 

```python
import ast 
```

```python
def clear(object):
    list=[]
    for i in ast.literal_eval(object):
        list.append(i['name'])
    return list
```

```python
final_data['genres']=final_data['genres'].apply(clear)
```

```python
final_data[['genres']].head()
```

```python
final_data['keywords']=final_data['keywords'].apply(clear)
final_data[['keywords']].head()
```

```python
# Extracting top 3 characters from the cast column

def top3_characters(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
final_data['cast']=final_data['cast'].apply(top3_characters)
final_data[['cast']].head()
```

```python
# Extracting director from the crew column 

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
final_data['crew']=final_data['crew'].apply(fetch_director)
final_data[['crew']].head()
```

```python
final_data['overview']=final_data['overview'].apply(lambda x:x.split())
```

```python
def transformation(list):
    l=[]
    for i in list:
        l.append(i.replace(" ",""))
    return l
```

```python
final_data['cast'] = final_data['cast'].apply(transformation)
final_data['crew'] = final_data['crew'].apply(transformation)
final_data['genres'] = final_data['genres'].apply(transformation)
final_data['keywords'] = final_data['keywords'].apply(transformation)
```

```python
final_data['tags'] = final_data['overview'] + final_data['genres'] + final_data['keywords'] + final_data['cast'] + final_data['crew']
```

```python
final_data.head()
```

```python
new_dataframe=final_data.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_dataframe.head()
```

```python
new_dataframe['tags'] = new_dataframe['tags'].apply(lambda x: " ".join(x))
new_dataframe.head()
```

## Text Vectorisation 

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=count_vectorizer.fit_transform(new_dataframe['tags']).toarray()
```

```python
vectors
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer=PorterStemmer()
```

```python
def stem(text):
    list=[]
    for i in text.split():
        list.append(porter_stemmer.stem(i))

    return "".join(list)
        
```

```python
new_dataframe['tags']=new_dataframe['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index=new_dataframe[new_dataframe['title']== movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_dataframe.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
